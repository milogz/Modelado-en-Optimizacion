# Modelado en Optimización (IIND-2501)

## Lección 3.4: '*Overview*' del Método Simplex

**Propósito.**
- Visualizar iteraciones del Método Simplex: base $B$, no-base $N$, multiplicadores, costos reducidos, **dirección** y **longitud de paso**.
- Desarrollar la intuición de *moverse de vértice en vértice* intercambiando variables básicas y no básicas.


## Modelo de programación lineal

Partimos de un modelo de programación lineal en formato estándar:

$$\max\; c^\top x$$
$$\text{s.a.}$$
$$A x = b,\;$$
$$x\ge 0.$$

Y lo separamos en partes **básica** y **no básica**:

$$\max\; c_B^\top x_B$$
$$\text{s.a.}$$
$$B x_B + N x_N = b,\;$$
$$x_B\ge 0,$$
$$x_N\ge 0.$$


## Pasos del método Simplex

1. Dada una **base factible inicial** ($I_B$ y $I_N$), podemos calcular el valor de las variables básicas y la función objetivo.

   $$x = \left[ \begin{array}{c} x_B \\ \hline x_N \end{array}\right] =
    \left[
    \begin{array}{c}
    B^{-1}b \\
    \hline
    0
    \end{array}
    \right], \quad \quad z=c_B^{T}x_B$$

2. **Verificar optimalidad**: calcular *costos reducidos*. (Recordar $w^{T} = c_B^{T} B^{-1} $)
   $$r = c_N - (c_B^{T} B^{-1}) N $$
   $$r = c_N - w^{T} N$$
   $$r_q = c_q - w^{T} a_q, \quad \forall q \in I_N$$
   
Si existe al menos un $r_q > 0$ para maximización (o un $r_q < 0$ para minimización), la variable no básica `q` **entra a la base** (pues su $r_q$ promete una mejora en la función objetivo); de lo contrario, decretamos optimalidad y terminamos.

3. **Definir dirección** (a partir de la variable entrante $q$, si la hay):

    $$d^{q} = \left[ \begin{array}{c} d^{q}_B \\ \hline d^{q}_N \end{array}\right] =
    \left[
    \begin{array}{c}
    -B^{-1} A_q \\
    \hline
    e_q
    \end{array}
    \right]$$

4. **Longitud**: encontramos cuánto podemos avanzar en la dirección $d^q$ sin que ninguna variable se haga negativa (hasta que alguna se haga cero).

$$
\theta = \min_{j\in I_B} \left\{ \frac{-x_{j}}{d_{j}} \;\middle|\; d_{j} < 0 \right\}$$



Si ninguna componente de la dirección es negativa, el problema es **no acotado**.

8. **Pivoteo**: actualizar base (sale la variable básica) y repetir.

$$x_{t+1} = x_{t} + \theta d^q$$

A continuación, un *widget* que deja **correr los pasos manualmente** y ver la geometría en 2D.

```

import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display, Math, HTML
import ipywidgets as W

# Import robusto de helpers (local ./lp_helpers o ./src o archivos sueltos)
import sys, os
from pathlib import Path

def _try_import():
    candidates = [
        Path.cwd() / "lp_helpers",
        Path.cwd() / "src",
        Path.cwd()
    ]
    for p in candidates:
        if p.exists() and str(p) not in sys.path:
            sys.path.append(str(p))
    try:
        from helpers_simplex import (
            mat_to_bmatrix, vec_to_bmatrix,
            basis_info, step_info,
        )
    except Exception as e:
        raise ImportError("No puedo importar helpers_simplex. Asegúrate de tener helpers_simplex.py accesible.") from e
    return dict(
        mat_to_bmatrix=mat_to_bmatrix, vec_to_bmatrix=vec_to_bmatrix,
        basis_info=basis_info, step_info=step_info
    )

H = _try_import()
mat_to_bmatrix = H['mat_to_bmatrix']
vec_to_bmatrix = H['vec_to_bmatrix']
basis_info = H['basis_info']
step_info = H['step_info']

# Config global de plots
plt.rcParams['figure.dpi'] = 120
```

```

def resumen_base_html(A, b, c, Bcols, var_names=None):
    \"\"\"
    Envuelve basis_info() y devuelve HTML con matrices y vectores clave.
    \"\"\"
    info = basis_info(A, b, c, Bcols)
    B, N = info["B"], info["N"]
    x, xB = info["x"], info["xB"]
    cB, cN = info["cB"], info["cN"]
    I_B = [j+1 for j,v in enumerate(Bcols) if v]
    I_N = [j+1 for j in info["Ncols"]]

    def _vec(v): return r"$" + vec_to_bmatrix(v) + r"$"
    def _mat(M): return r"$" + mat_to_bmatrix(M) + r"$"

    html = f\"\"\"
    <h4>Base actual</h4>
    <p><b>B (índices 1-based):</b> {I_B}&nbsp;&nbsp;&nbsp;<b>N:</b> {I_N}</p>
    <p><b>x</b> = {_vec(x)}</p>
    <p><b>c_B</b> = {_vec(cB)}&nbsp;&nbsp;&nbsp;<b>c_N</b> = {_vec(cN)}</p>
    <p><b>B</b> = {_mat(B)}&nbsp;&nbsp;&nbsp;<b>N</b> = {_mat(N)}</p>
    \"\"\"
    return info, HTML(html)

def mover_un_paso_html(A, b, c, Bcols, enter_idx):
    \"\"\"
    Aplica step_info() con columna entrante enter_idx (0-based) y devuelve
    dict con detalles + HTML de resumen.
    \"\"\"
    info = basis_info(A, b, c, Bcols)
    x = info["x"]
    step = step_info(A, x, Bcols, enter_idx)
    d, dB, theta, leave, x_new = step["d"], step["dB"], step["theta"], step["leave"], step["x_new"]

    def _vec(v): return r"$" + vec_to_bmatrix(v) + r"$"

    if np.isfinite(theta) and (leave is not None):
        msg = f"θ = {theta:g}, sale columna (0-based) {leave}."
    else:
        msg = "Dirección no mejora (no acotado o sin razón de salida)."

    html = f\"\"\"
    <h4>Movimiento</h4>
    <p><b>d</b> = {_vec(d)}</p>
    <p><b>d_B</b> = {_vec(dB)}</p>
    <p>{msg}</p>
    <p><b>x nuevo</b> = {_vec(x_new)}</p>
    \"\"\"
    return step, HTML(html)
```

```

# Ejemplo didáctico (reemplaza por el que ya usas en el notebook original si lo deseas)
A = np.array([
    [1, 2, 1, 0, 0],
    [2, 1, 0, 1, 0],
    [0, 1, 0, 0, 1],
], dtype=float)

b = np.array([4, 4, 2], dtype=float)
c = np.array([3, 2, 0, 0, 0], dtype=float)

# Base inicial: slacks activos (s1,s2,s3) -> columnas 3,4,5 (0-based)
Bcols = [False, False, True, True, True]  # x1,x2,s1,s2,s3
```

```

log = W.Output()
log2 = W.Output()

btn_info = W.Button(description="1) Info Base", button_style="primary")
btn_step = W.Button(description="2) Mover (entra j)")

def _options_nonbasic(Bcols):
    return [i for i,v in enumerate(Bcols) if not v]

drop_enter = W.Dropdown(options=_options_nonbasic(Bcols), value=_options_nonbasic(Bcols)[0], description="Entra j")

def _on_info_clicked(_):
    with log:
        log.clear_output()
        info, html = resumen_base_html(A,b,c,Bcols)
        display(html)

def _on_step_clicked(_):
    global Bcols
    j = drop_enter.value
    step, html = mover_un_paso_html(A,b,c,Bcols,j)
    with log2:
        log2.clear_output()
        display(html)
    leave = step["leave"]
    if np.isfinite(step["theta"]) and leave is not None:
        Bcols[leave] = False
        Bcols[j] = True
        drop_enter.options = _options_nonbasic(Bcols)

btn_info.on_click(_on_info_clicked, remove=True); btn_info.on_click(_on_info_clicked)
btn_step.on_click(_on_step_clicked, remove=True); btn_step.on_click(_on_step_clicked)

ui = W.VBox([W.HBox([btn_info, drop_enter, btn_step]), log, log2])
display(ui)
```
