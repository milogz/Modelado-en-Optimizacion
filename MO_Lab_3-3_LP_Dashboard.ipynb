### Conceptos de referencia rápida

**Formato estándar** de un LP:

$$\max\ c^T x$$
$$s.a:$$
$$Ax + s = b$$
$$x \ge 0$$
$$s \ge 0$$

| Concepto | Significado | Interpretación geométrica |
|---|---|---|
| **Holgura** $s_i = b_i - a_i^T x$ | Cuánto 'le sobra' a la restricción $i$ | Distancia del punto a la frontera de la restricción |
| **Restricción activa** ($s_i = 0$) | Restricción satisfecha con igualdad | El punto está **sobre** la recta de la restricción |
| **Base** $B$ | Subconjunto de $m$ columnas de $[A|I]$ | Intersección de $n$ restricciones activas → vértice |
| **BFS** (sol. básica factible) | $x_B = B^{-1}b \ge 0$ | Vértice del poliedro factible |
| **Costo reducido** $\bar{c}_j$ | Tasa de mejora si $x_j$ entrara a la base | ¿Conviene moverse en esa dirección? |

### Problema ejemplo

$$\max\ z = 3x_1 + 2x_2$$

Sujeto a:
$$
\begin{aligned}
x_1 + x_2 &\le 6 \\
x_1 + 2x_2 &\le 8 \\
2x_1 + x_2 &\le 8 \\
x_1, x_2 &\ge 0
\end{aligned}
$$

**Formato estándar** (con holguras $s_1, s_2, s_3$):

$$
[A|I] = \begin{bmatrix} 1 & 1 & 1 & 0 & 0 \\ 1 & 2 & 0 & 1 & 0 \\ 2 & 1 & 0 & 0 & 1 \end{bmatrix}, \quad
b = \begin{bmatrix} 6 \\ 8 \\ 8 \end{bmatrix}, \quad
c = \begin{bmatrix} 3 \\ 2 \\ 0 \\ 0 \\ 0 \end{bmatrix}
$$

```python
A = np.array([[1.,1.],[1.,2.],[2.,1.]])
b = np.array([6.,8.,8.])
c = np.array([3.,2.])

lp_dashboard(A, b, c)
```
